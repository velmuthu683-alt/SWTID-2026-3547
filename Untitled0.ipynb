```python
import json
import json5
import gradio as gr
from enum import Enum
from jsonschema import validate
from pydantic import BaseModel
```

```python
class CaseStyle(Enum):
    SNAKE = "snake_case"
    CAMEL = "camelCase"
    KEBAB = "kebab-case"
    PASCAL = "PascalCase"
```

```python
class JSONRefinerCore:

    def __init__(self):
        self.errors = []
```

```python
def infer_type(self, value):

    if value.lower() in ["true","yes","on"]:
        return True

    if value.lower() in ["false","no","off"]:
        return False

    if value.lower() in ["null","none","nil","n/a"]:
        return None

    try:
        if "." in value:
            return float(value)
        return int(value)
    except:
        return value
```

```python
def to_snake_case(self,key):
    return key.lower().replace(" ","_")
```

```python
def to_camel_case(self,key):
    words = key.split(" ")
    return words[0].lower() + "".join(w.capitalize() for w in words[1:])
```

```python
def to_kebab_case(self,key):
    return key.lower().replace(" ","-")
```

```python
def to_pascal_case(self,key):
    return "".join(word.capitalize() for word in key.split(" "))
```

```python
def normalize_key(self,key,style):

    if style == "snake_case":
        return self.to_snake_case(key)

    if style == "camelCase":
        return self.to_camel_case(key)

    if style == "kebab-case":
        return self.to_kebab_case(key)

    if style == "PascalCase":
        return self.to_pascal_case(key)

    return key
```

```python
def validate_json_schema(self,data,schema):

    try:
        validate(instance=data, schema=schema)
        return "Schema Valid"

    except Exception as e:
        return str(e)
```

```python
def check_required_fields(self,data,required):

    missing = []

    for field in required:
        if field not in data:
            missing.append(field)

    return missing
```

```python
def flatten_json(self,data,parent_key="",sep="."):

    items = {}

    for k,v in data.items():

        new_key = parent_key + sep + k if parent_key else k

        if isinstance(v,dict):
            items.update(self.flatten_json(v,new_key,sep))

        else:
            items[new_key] = v

    return items
```

```python
def remove_null_values(self,data):

    return {k:v for k,v in data.items() if v is not None}
```

```python
def refine_json(input_text,case_style):

    core = JSONRefinerCore()

    try:
        data = json5.loads(input_text)

        result = {}

        for k,v in data.items():

            new_key = core.normalize_key(k,case_style)

            result[new_key] = core.infer_type(str(v))

        return json.dumps(result,indent=4)

    except Exception as e:
        return str(e)
```

```python
import json
import json5

class JSONRefinerCore:

    def __init__(self):
        self.errors = []

    # Type inference
    def infer_type(self, value):

        if value.lower() in ["true","yes","on"]:
            return True

        if value.lower() in ["false","no","off"]:
            return False

        if value.lower() in ["null","none","nil","n/a"]:
            return None

        try:
            if "." in value:
                return float(value)
            return int(value)
        except:
            return value


    # Snake case
    def to_snake_case(self, key):
        return key.lower().replace(" ", "_")


    # Camel case
    def to_camel_case(self, key):
        words = key.split(" ")
        return words[0].lower() + "".join(w.capitalize() for w in words[1:])


    # Kebab case
    def to_kebab_case(self, key):
        return key.lower().replace(" ", "-")


    # Pascal case
    def to_pascal_case(self, key):
        return "".join(word.capitalize() for word in key.split(" "))


    # KEY NORMALIZATION FUNCTION
    def normalize_key(self, key, style):

        if style == "snake_case":
            return self.to_snake_case(key)

        elif style == "camelCase":
            return self.to_camel_case(key)

        elif style == "kebab-case":
            return self.to_kebab_case(key)

        elif style == "PascalCase":
            return self.to_pascal_case(key)

        return key
```

```python
def refine_json(input_text, case_style):

    core = JSONRefinerCore()

    try:
        data = json5.loads(input_text)

        result = {}

        for k, v in data.items():

            new_key = core.normalize_key(k, case_style)

            result[new_key] = core.infer_type(str(v))

        return json.dumps(result, indent=4)

    except Exception as e:
        return str(e)
```

```python
import gradio as gr

with gr.Blocks() as demo:

    gr.Markdown("# JSON Refiner Tool")

    input_box = gr.Textbox(
        label="Enter JSON Input",
        lines=10
    )

    case_style = gr.Radio(
        ["snake_case", "camelCase", "kebab-case", "PascalCase"],
        label="Select Key Case Style"
    )

    output_box = gr.Code(
        label="Refined JSON Output",
        language="json"
    )

    refine_button = gr.Button("Refine JSON")

    refine_button.click(
        refine_json,
        inputs=[input_box, case_style],
        outputs=output_box
    )

demo.launch()
```
